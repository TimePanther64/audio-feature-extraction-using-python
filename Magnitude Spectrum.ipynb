```python
import numpy as np
import matplotlib.pyplot as plt
import os
import IPython.display as ipd
import librosa
```

```python
violin_file = 'audio/violin_c.wav'
sax_file = 'audio/sax.wav'
piano_file = 'audio/piano_c.wav'
noise_file = 'audio/noise.wav'
```

```python
violin, sr = librosa.load(violin_file)
sax, _ =  librosa.load(sax_file)
piano, _ = librosa.load(piano_file)
noise, _ = librosa.load(noise_file)
```

```python
def plot_ms(signal, title, sr, f_ratio=1):
    ft = np.fft.fft(signal)
    ms = np.abs(ft)
    freq = np.linspace(0, sr, len(ms))
    k = len(freq)
    
    plt.figure(figsize=(15,10))
    plt.plot(freq[:int(k*f_ratio)], ms[:int(k*f_ratio)])
    plt.xlabel('Frequency')
    plt.ylabel('Magnitude Spectrum')
    plt.title(title)
    
    plt.show()

# plot_ms(sax, title='Sax', sr=sr, f_ratio=0.2)
# plot_ms(violin, title='Violin', sr=sr, f_ratio=0.2)
# plot_ms(piano, title='Piano', sr=sr, f_ratio=0.2)
plot_ms(noise, title='Noise', sr=sr, f_ratio=.1)
```
