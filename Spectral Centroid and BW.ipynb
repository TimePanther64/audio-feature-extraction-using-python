```python
import math
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd 
import librosa
import librosa.display
```

```python
debussy_file = 'audio/debussy.wav'
redhot_file = 'audio/redhot.wav'
duke_file = 'audio/duke.wav'

debussy, sr = librosa.load(debussy_file)
redhot, _  = librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)
```

```python
FRAME_SIZE = 2048
HOP_SIZE = 512

debussy_spec = librosa.stft(debussy, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
redhot_spec = librosa.stft(redhot, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
duke_spec = librosa.stft(duke, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
```

```python
sc_debussy = librosa.feature.spectral_centroid(y=debussy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
sc_redhot = librosa.feature.spectral_centroid(y=redhot, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
sc_duke = librosa.feature.spectral_centroid(y=duke, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
frame = range(len(sc_debussy))
t = librosa.frames_to_time(frame, hop_length=HOP_SIZE)

plt.figure(figsize=(15,10))
plt.plot(t, sc_debussy, color='b')
plt.plot(t, sc_duke, color='g')
plt.plot(t, sc_redhot, color='r')
plt.show()
```

```python
bw_debussy = librosa.feature.spectral_bandwidth(y=debussy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
bw_redhot = librosa.feature.spectral_bandwidth(y=redhot, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
bw_duke = librosa.feature.spectral_bandwidth(y=duke, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
frame = range(len(sc_debussy))
t = librosa.frames_to_time(frame, hop_length=HOP_SIZE)

plt.figure(figsize=(15,10))
plt.plot(t, bw_debussy, color='b')
plt.plot(t, bw_duke, color='g')
plt.plot(t, bw_redhot, color='r')
plt.show()
```
